# Summer Olympics Data Analysis Assignment

```python
import pandas as pd

df = pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
most_golds = df[df["Medal"]=="Gold"].groupby("Sport")["Medal"].agg('size').to_frame("Golds").sort_values(by="Golds",ascending=False).reset_index().head()
most_golds.plot(x = 'Sport', y = 'Golds', kind = 'bar', figsize = (9,6))
print(most_golds)
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
most_medals = df.groupby("Sport").Medal.agg('size').to_frame("Golds").sort_values(by="Golds",ascending=False).reset_index().head()
most_medals.plot(x = 'Sport', y = 'Golds', kind = 'bar', figsize = (9,6))
print(most_medals)
```

### 4. Which player has won most number of medals? (Top 5)

```python
player_most_medals = df.groupby("Athlete").Medal.agg('size').to_frame("Medals").sort_values(by="Medals",ascending=False).reset_index().head()
player_most_medals.plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (9,6))
print(player_most_medals)
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = df[df["Medal"]=="Gold"].groupby("Athlete")["Medal"].agg('size').to_frame("Golds").sort_values(by="Golds",ascending=False).reset_index().head()
data.plot(x = 'Athlete', y = 'Golds', kind = 'bar', figsize = (9,6))
print(data)
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = df[(df["Country"]=="IND")&(df["Medal"]=="Gold")].sort_values(by="Medal",ascending=False).reset_index()
data.Year.iloc[0]
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = df.groupby("Event")["Athlete"].agg('size').to_frame("Athlete").sort_values(by="Athlete",ascending=False).reset_index().head()
data.plot(x = 'Event', y = 'Athlete', kind = 'bar', figsize = (9,6))
print(data)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = df[(df["Gender"]=="Women")&(df["Medal"]=="Gold")].groupby("Sport").Medal.agg("size").to_frame("Gold Medals").sort_values(by="Gold Medals",ascending=False).reset_index().head()
data.plot(x = 'Sport', y = 'Gold Medals', kind = 'bar', figsize = (9,6))
print(data)
```
